# U.S. Medical Insurance Costs

```python
# Importing libraries
import csv
from scipy.stats import pearsonr
```

```python
# Average insurance cost by age

csv_file = open('insurance.csv')
csv_reader = csv.reader(csv_file, delimiter=',')
ages = []
insurance_costs = []
for row in csv_reader:
  ages.append(row[0])
  insurance_costs.append(row[6])

ages.pop(0) # remove the header
insurance_costs.pop(0)

ages = [int(age) for age in ages]
insurance_costs = [float(cost) for cost in insurance_costs]

total_cost = 0
for cost in insurance_costs:
  total_cost += cost

average_cost = total_cost / len(insurance_costs)
print("Average Insurance Cost: " + str(average_cost) + " dollars.")

# Average insurance cost by age
costs_by_age = {}
for i in range(len(ages)):
    costs_by_age[ages[i]] = insurance_costs[i]
costs_by_age = dict(sorted(costs_by_age.items()))

print(costs_by_age) 
```

```python
# Average Cost by Gender
import csv

csv_file = open('insurance.csv')
csv_reader = csv.reader(csv_file, delimiter=',')

genders = []
insurance_costs = []
for row in csv_reader:
    genders.append(row[1])
    insurance_costs.append(row[6])

genders.pop(0)
insurance_costs.pop(0)

insurance_costs = [float(cost) for cost in insurance_costs]

total_cost = 0
for cost in insurance_costs:
    total_cost += cost
average_cost = total_cost / len(insurance_costs)
print("Average Insurance Cost: " + str(average_cost) + " dollars.")

# Average insurance cost by gender
genders_costs = zip(genders, insurance_costs)
# print(list(genders_costs))

male_costs = []
female_costs = []
for i in genders_costs:
    if i[0] == 'female':
        female_costs.append(i[1])
    else:
        male_costs.append(i[1])

# print(male_costs)
# print(female_costs)

#male average cost
male_total = 0
for cost in male_costs:
    male_total += cost

average_male_cost = male_total / len(male_costs)
print("Average Insurace Cost for Males: " + str(average_male_cost) + " dollars.")

#female average cost
female_total = 0
for cost in female_costs:
    female_total += cost

average_female_cost = female_total / len(female_costs)
print("Average Insurance Cost for Females: " + str(average_female_cost) + " dollars.")
```

```python
# Insurance cost by BMI
import csv

csv_file = open('insurance.csv')
csv_reader = csv.reader(csv_file, delimiter=',')

bmis = []
insurance_costs = []

for row in csv_reader:
    bmis.append(row[2])
    insurance_costs.append(row[6])

bmis.pop(0)
insurance_costs.pop(0)

insurance_costs = [float(cost) for cost in insurance_costs] 
bmis = [float(bmi) for bmi in bmis]

# Average insurance cost by BMI
bmi_costs = zip(bmis, insurance_costs)
# print(list(bmi_costs))

underweight_costs = []
normal_costs = []
overweight_costs = []
obese_costs = []

for i in bmi_costs:
    if i[0] < 18.5:
        underweight_costs.append(i[1])
    elif i[0] >= 18.5 and i[0] <= 24.9:
        normal_costs.append(i[1])
    elif i[0] >= 25 and i[0] <= 29.9:
        overweight_costs.append(i[1])
    else:
        obese_costs.append(i[1])

# print(underweight_costs)
# print(normal_costs)
# print(overweight_costs)
# print(obese_costs)

# underweight average cost
underweight_total = 0
for cost in underweight_costs:
    underweight_total += cost

average_underweight_cost = underweight_total / len(underweight_costs)
print("Average Insurance Cost for Underweight: " + str(average_underweight_cost) + " dollars.")

# normal average cost
normal_total = 0
for cost in normal_costs:
    normal_total += cost

average_normal_cost = normal_total / len(normal_costs)
print("Average Insurance Cost for Normal: " + str(average_normal_cost) + " dollars.")

# overweight average cost
overweight_total = 0
for cost in overweight_costs:
    overweight_total += cost

average_overweight_cost = overweight_total / len(overweight_costs)
print("Average Insurance Cost for Overweight: " + str(average_overweight_cost) + " dollars.")

# obese average cost
obese_total = 0
for cost in obese_costs:
    obese_total += cost
    
average_obese_cost = obese_total / len(obese_costs)
print("Average Insurance Cost for Obese: " + str(average_obese_cost) + " dollars.")
```

```python
# Insurance Cost by Number of Children

import csv

csv_file = open('insurance.csv')
csv_reader = csv.reader(csv_file, delimiter=',')

children = []
insurance_costs = []

for row in csv_reader:
    children.append(row[3])
    insurance_costs.append(row[6])

children.pop(0)
insurance_costs.pop(0)

insurance_costs = [float(cost) for cost in insurance_costs]
children = [int(child) for child in children]

# Average insurance cost by number of children
children_costs = zip(children, insurance_costs)
# print(list(children_costs))

zero_costs = []
one_costs = []
two_costs = []
three_costs = []
four_costs = []
five_costs_or_more = []

for i in children_costs:
    if i[0] == 0:
        zero_costs.append(i[1])
    elif i[0] == 1:
        one_costs.append(i[1])
    elif i[0] == 2:
        two_costs.append(i[1])
    elif i[0] == 3:
        three_costs.append(i[1])
    elif i[0] == 4:
        four_costs.append(i[1])
    else:
        five_costs_or_more.append(i[1])

# print(zero_costs)
# print(one_costs)
# print(two_costs)
# print(three_costs)
# print(four_costs)
# print(five_costs_or_more)

# zero average cost
zero_total = 0
for cost in zero_costs:
    zero_total += cost

average_zero_cost = zero_total / len(zero_costs)
print("Average Insurance Cost for Zero Children: " + str(average_zero_cost) + " dollars.")

# one average cost
one_total = 0
for cost in one_costs:
    one_total += cost

average_one_cost = one_total / len(one_costs)
print("Average Insurance Cost for One Child: " + str(average_one_cost) + " dollars.")

# two average cost
two_total = 0
for cost in two_costs:
    two_total += cost

average_two_cost = two_total / len(two_costs)
print("Average Insurance Cost for Two Children: " + str(average_two_cost) + " dollars.")

# three average cost
three_total = 0
for cost in three_costs:
    three_total += cost

average_three_cost = three_total / len(three_costs)
print("Average Insurance Cost for Three Children: " + str(average_three_cost) + " dollars.")

# four average cost
four_total = 0
for cost in four_costs:
    four_total += cost

average_four_cost = four_total / len(four_costs)
print("Average Insurance Cost for Four Children: " + str(average_four_cost) + " dollars.")

# five or more average cost
five_total = 0
for cost in five_costs_or_more:
    five_total += cost

average_five_cost = five_total / len(five_costs_or_more)
print("Average Insurance Cost for Five or More Children: " + str(average_five_cost) + " dollars.")
```

```python
# Smoker vs. Non-Smoker Insurance Costs
import csv
csv_file = open('insurance.csv')
csv_reader = csv.reader(csv_file, delimiter=',')

smoker = []
insurance_costs = []

for row in csv_reader:
    smoker.append(row[4])
    insurance_costs.append(row[6])

smoker.pop(0)
insurance_costs.pop(0)

insurance_costs = [float(cost) for cost in insurance_costs]

# Average insurance cost by smoker
smoking_costs = zip(smoker, insurance_costs)
# print(list(smoking_costs))

smoker_costs = []
non_smoker_costs = []

for i in smoking_costs:
    if i[0] == 'yes':
        smoker_costs.append(i[1])
    else:
        non_smoker_costs.append(i[1])

# smoker average cost
smoker_total = 0
for cost in smoker_costs:
    smoker_total += cost

average_smoker_cost = smoker_total / len(smoker_costs)
print("Average Insurance Cost for Smokers: " + str(average_smoker_cost) + " dollars.")

# non-smoker average cost
non_smoker_total = 0
for cost in non_smoker_costs:
    non_smoker_total += cost

average_non_smoker_cost = non_smoker_total / len(non_smoker_costs)
print("Average Insurance Cost for Non-Smokers: " + str(average_non_smoker_cost) + " dollars.")
```

```python
# Correlation between Age and Insurance Costs
from scipy.stats import pearsonr
import csv
csv_file = open('insurance.csv')
csv_reader = csv.reader(csv_file, delimiter=',')

age = []
insurance_costs = []

for row in csv_reader:
    age.append(row[0])
    insurance_costs.append(row[6])

age.pop(0)
insurance_costs.pop(0)

insurance_costs = [float(cost) for cost in insurance_costs]
age = [int(age) for age in age]

def calculate_correlation(list1, list2):
    correlation_coeff, _ = pearsonr(list1, list2)
    return correlation_coeff

age_cost_correlation = calculate_correlation(age, insurance_costs)
print("Pearson Correlation between Age and Insurance Costs: " + str(age_cost_correlation))

# Correlation between BMI and Insurance Costs
from scipy.stats import pearsonr
import csv
csv_file = open('insurance.csv')
csv_reader = csv.reader(csv_file, delimiter=',')

bmis = []
insurance_costs = []

for row in csv_reader:
    bmis.append(row[2])
    insurance_costs.append(row[6])

bmis.pop(0)
insurance_costs.pop(0)

insurance_costs = [float(cost) for cost in insurance_costs]
bmis = [float(bmi) for bmi in bmis]

bmi_cost_correlation = calculate_correlation(bmis, insurance_costs)
print("Pearson Correlation between BMI and Insurance Costs: " + str(bmi_cost_correlation))
```
